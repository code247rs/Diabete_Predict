```python
import pandas as pd
from sklearn.preprocessing import StandardScaler


df= pd.read_csv('Dataset of Diabetes .csv')

df=df.dropna(thresh=10)
#df=df.fillna(df.mean(numeric_only=True), inplace=True)
df.head()
cols_to_scale = ['AGE', 'Urea', 'Cr', 'HbA1c', 'Chol', 'TG', 'HDL', 'LDL', 'VLDL', 'BMI']

stdScaler= StandardScaler()

df[cols_to_scale]=stdScaler.fit_transform(df[cols_to_scale])
df.head()

df['Gender']= df['Gender'].apply(lambda x: 0 if x=='M' else 1)
df['CLASS'] = df['CLASS'].map({'N': 0, 'Pre-diabetic': 1, 'Y': 2})

   

df.head()

clean_data= df.drop(columns=['ID', 'No_Pation'])

clean_data.head()
clean_data.iloc[230:300]

```
